```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the differential amplifier
V_max = 5.0  # Maximum output voltage in volts
A_v = 400.0  # Voltage gain (from previous calculations)

# Create a range of differential input voltages
diff_input_voltages = np.linspace(-10e-3, 10e-3, 1000)  # -10mV to 10mV

# Calculate the output voltage
# Using a simplified linear model for the differential amplifier
output_voltages = A_v * diff_input_voltages
# Clipping the output to the maximum voltage
output_voltages = np.clip(output_voltages, -V_max, V_max)

# Plotting the voltage transfer characteristic
plt.figure(figsize=(10, 6))
plt.plot(diff_input_voltages * 1000, output_voltages, label='Voltage Transfer Characteristic')
plt.xlabel('Differential Input Voltage (mV)')
plt.ylabel('Output Voltage (V)')
plt.title('Voltage Transfer Characteristic of a Differential Amplifier')
plt.grid(True)
plt.legend()
plt.show()

```
