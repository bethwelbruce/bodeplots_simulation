```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Operational Amplifier Parameters
gain = 100000  # Open-loop gain
f_c = 10  # Cutoff frequency in Hz (Change this as per your op-amp specs)

# Transfer Function (Low-pass filter approximation)
s1 = signal.TransferFunction([gain], [1/(2*np.pi*f_c), 1])

# Frequency range for Bode plot
frequencies = np.logspace(1, 6, 500)  # 10 Hz to 1 MHz

# Generate Bode plot
w, mag, phase = signal.bode(s1, frequencies)

# Plotting
plt.figure(figsize=(12, 8))

# Gain Plot
plt.subplot(2, 1, 1)
plt.semilogx(w/(2*np.pi), mag)  # Convert to Hz for frequency
plt.title('Bode Plot of an Operational Amplifier')
plt.ylabel('Gain (dB)')
plt.grid(which='both', axis='both')

# Phase Plot
plt.subplot(2, 1, 2)
plt.semilogx(w/(2*np.pi), phase)  # Convert to Hz for frequency
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.grid(which='both', axis='both')

plt.show()
```
