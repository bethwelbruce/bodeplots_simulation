```python
import numpy as np
import matplotlib.pyplot as plt

# Component values (from previous calculations)
R1 = R2 = 10e3  # ohms
RE = 1e3       # ohms
I = 1e-3       # amps
beta = 100
Vt = 26e-3     # volts

# Calculate the emitter resistance (re)
re = 1 / (0.026 * I)

# Calculate the transconductance (gm)
gm = beta * I / Vt

# Calculate the voltage gain (Av_diff)
Av_diff = - (R2 / R1) * (gm * RE)

# Create a range of input voltage values
Vin_diff = np.linspace(-50e-3, 50e-3, 1000)

# Calculate the output voltage based on the theoretical model
Vout = -Av_diff * Vin_diff

# Plot the voltage transfer characteristic
plt.plot(Vin_diff, Vout)
plt.xlabel("Differential Input Voltage (V)")
plt.ylabel("Output Voltage (V)")
plt.title("Voltage Transfer Characteristic of Differential Amplifier Input Stage")
plt.grid(True)
plt.show()
```
