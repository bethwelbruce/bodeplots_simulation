```python
import numpy as np
import matplotlib.pyplot as plt

# Frequency range for the Bode plots
frequencies = np.logspace(1, 6, 500)  # 10 Hz to 1 MHz

# Op-amp parameters
gain_dc = 1e5   # DC gain (100,000 or 100 dB)
f_pole = 10e3   # Frequency of the dominant pole (10 kHz)

# Gain and phase calculations
gain = gain_dc / np.sqrt(1 + (frequencies / f_pole)**2)
phase = -np.arctan(frequencies / f_pole) * (180 / np.pi)

# Plotting Gain Magnitude vs. Frequency
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.semilogx(frequencies, 20 * np.log10(gain))
plt.title('Bode Plot of an Operational Amplifier')
plt.ylabel('Gain (dB)')
plt.grid(which='both', axis='both')

# Plotting Phase vs. Frequency
plt.subplot(2, 1, 2)
plt.semilogx(frequencies, phase)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.grid(which='both', axis='both')
plt.show()
```
